```python
import pandas as pd
import csv
```

```python
tmdb_5000_cast = pd.read_csv('../tmdb_5000_cast.csv')
tmdb_5000_cast.head()
```

```python
tmdb_top_cast = tmdb_5000_cast.sort_values('order_score', ascending=False)[['cast_id', 'cast_name', 'movie_id']].head(5)
tmdb_top_cast
```

```python
tmdb_5000_movies = pd.read_csv('../tmdb_5000_movies_cleaned.csv')
tmdb_5000_movies = tmdb_5000_movies[[ 'id', 'original_title', 'popularity', 'title', 'vote_average', 'ROI']]
tmdb_5000_movies
```

```python
movie_cast_dict = {}
for i in tmdb_top_cast.index:
    mov_ids = tmdb_top_cast['movie_id'][i].split('_')
    cast_id = str(tmdb_top_cast['cast_name'][i]).replace('.', '')
    for mov_id in mov_ids:
        mov_id = int(mov_id)
        if mov_id in movie_cast_dict:
            movie_cast_dict[mov_id].append(cast_id)
        else:
            movie_cast_dict[mov_id] = [cast_id]
movie_cast_dict
```

```python
tmdb_5000_movies = tmdb_5000_movies.loc[tmdb_5000_movies['id'].isin(movie_cast_dict)]
tmdb_5000_movies['cast_names'] =  ''
for i in tmdb_5000_movies.index:
    mov_id = tmdb_5000_movies['id'][i]
    tmdb_5000_movies['cast_names'][i] = str('\"' + ','.join(movie_cast_dict[mov_id]) + '\"')
tmdb_5000_movies
```

```python

tmdb_5000_movies.to_csv('../top_cast.csv', quoting=csv.QUOTE_NONE, escapechar='\\' )
```
